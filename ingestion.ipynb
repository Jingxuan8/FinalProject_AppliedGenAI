```python
from pathlib import Path
import numpy as np
import pandas as pd
from datasets import load_dataset
```

```python
ds = load_dataset("calmgoose/amazon-product-data-2020", split="train")
df_raw = ds.to_pandas()
print(df_raw.shape)
df_raw.head()
```

```python
# Category = "Toys & Games | Games & Accessories"
PREFIX = "Toys & Games | Games & Accessories"

mask = df_raw["Category"].astype(str).str.startswith(PREFIX)
df_slice = df_raw[mask].copy()

print("Selected Games & Accessories slice rows:", len(df_slice))
df_slice["Category"].value_counts()
```

## Standardization & Cleaning

#### Standardization

```python
raw_cols = [
    "Uniq Id",
    "Product Name",
    "Category",
    "Selling Price",
    "About Product",
    "Product Specification",
    "Technical Details",
    "Shipping Weight",
    "Product Dimensions",
    "Product Url",
]

df_slice = df_slice[raw_cols].rename(columns=lambda c: c.strip())

df_slice.head()
```

```python
# 1) Basic mapping into a standardized schema
df_std = pd.DataFrame()
df_std["id"] = df_slice["Uniq Id"]
df_std["title"] = df_slice["Product Name"]
df_std["category"] = df_slice["Category"]
df_std["price_raw"] = df_slice["Selling Price"]

# 2) Merge text fields into a single "features" column
text_cols = ["About Product", "Product Specification", "Technical Details"]
for col in text_cols:
    df_slice[col] = df_slice[col].fillna("").astype(str)

def build_features(row):
    parts = [row[c].strip() for c in text_cols if row[c] and row[c].strip()]
    return " | ".join(parts)

df_std["features"] = df_slice.apply(build_features, axis=1)

# 3) Approximate brand: use the first token of the title
def extract_brand_from_title(title: str) -> str:
    if not isinstance(title, str) or not title.strip():
        return ""
    return title.strip().split()[0]

df_std["brand"] = df_std["title"].astype(str).apply(extract_brand_from_title)

# 4) rating / ingredients: keep as placeholders
df_std["rating"] = np.nan
df_std["ingredients"] = ""

# 5) Keep some additional useful metadata: weight / dimensions / URL
df_std["shipping_weight"] = df_slice["Shipping Weight"]
df_std["product_dimensions"] = df_slice["Product Dimensions"]
df_std["product_url"] = df_slice["Product Url"]

df_std.head()
```

#### Cleaning

```python
# Inspect missing values before dropping anything
key_cols = ["id", "title", "price_raw"]

print("Missing rate for key fields (before dropping):")
print(df_std[key_cols].isna().mean())

# Optional: look at counts as well
print("\nMissing counts for key fields (before dropping):")
print(df_std[key_cols].isna().sum())

# Now safely drop rows where any of the key fields are missing: id / title / price_raw
df_clean = df_std.dropna(subset=key_cols).copy()
print("\nRows after dropping rows with missing key fields:", len(df_clean))


def clean_price(x):
    """Convert raw price strings like '$19.99' into float values."""
    try:
        s = str(x)
        s = s.replace("$", "").replace(",", "").strip()
        return float(s)
    except Exception:
        return np.nan

# Parse price into a numeric column
df_clean["price"] = df_clean["price_raw"].apply(clean_price)

# Check price parse success rate
print("\nPrice parse failure rate:", df_clean["price"].isna().mean())

# Drop rows where price could not be parsed
df_clean = df_clean.dropna(subset=["price"])

# Remove obviously unreasonable prices (e.g., < 2 or > 300 for games).
# You can adjust these thresholds based on df_clean[['price']].describe().
df_clean = df_clean[df_clean["price"].between(2, 300)]
print("After price cleaning:", df_clean.shape)

# Quick sanity check of the remaining price distribution
df_clean[["price"]].describe()
```

```python
# Normalize weight
import re
def parse_weight_to_kg(w):
    if not isinstance(w, str):
        return np.nan
    s = w.lower()
    m = re.search(r"(\d+(\.\d+)?)", s)
    if not m:
        return np.nan
    qty = float(m.group(1))

    if "kilogram" in s or "kg" in s:
        return qty
    if "pound" in s or "lb" in s:
        return qty * 0.45359237
    if "ounce" in s or "oz" in s:
        return qty * 0.0283495
    return np.nan

df_clean["weight_kg"] = df_clean["shipping_weight"].apply(parse_weight_to_kg)

df_clean["price_per_kg"] = df_clean.apply(
    lambda r: r["price"] / r["weight_kg"]
    if pd.notnull(r.get("weight_kg", np.nan)) and r["weight_kg"] > 0
    else np.nan,
    axis=1,
)

df_clean[["price", "shipping_weight", "weight_kg", "price_per_kg"]].head()
```

## Save Cleaned Dataset

```python
DATA_DIR = Path("data")
PROCESSED_DIR = DATA_DIR / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

PROCESSED_PATH = PROCESSED_DIR / "games_accessories.parquet"

final_cols = [
    "id",
    "title",
    "brand",
    "category",
    "price",
    "rating",
    "features",
    "ingredients",
    "weight_kg",
    "price_per_kg",
    "product_url",
]

df_final = df_clean[final_cols].drop_duplicates(subset=["id"]).reset_index(drop=True)

print("Final processed shape:", df_final.shape)
df_final.head()
```

```python
df_final.to_parquet(PROCESSED_PATH, index=False)
df_final.to_csv(PROCESSED_DIR / "games_accessories.csv", index=False)

print("Saved processed subset to:", PROCESSED_PATH)
```
